```python
from bs4 import BeautifulSoup

import requests

# Function to extract Product Title

def get_title(soup):

  try:

    title = soup.find("span", attrs={"id": "productTitle"})

    title_string = title.get_text(strip=True)

  except AttributeError:

    title_string = ""

  return title_string



# Function to extract Product Price

def get_price(soup):

  try:

    price = soup.find("span", attrs={"class": "a-offscreen"})

    if price is not None:

      price_string = price.get_text(strip=True)

    else:

      price_string = ""

  except AttributeError:

    price_string = ""

  return price_string



# Function to extract Product Rating

def get_rating(soup):

  try:

    rating = soup.find("span", attrs={"class": "a-icon-alt"})

    if rating is not None:

      rating_string = rating.get_text(strip=True)

    else:

      rating_string = ""

  except AttributeError:

    rating_string = ""

  return rating_string



# Function to extract Number of User Reviews

def get_review_count(soup):

  try:

    review_count = soup.find("span", attrs={"id": "acrCustomerReviewText"})

    if review_count is not None:

      review_count_string = review_count.get_text(strip=True)

    else:

      review_count_string = ""

  except AttributeError:

    review_count_string = ""

  return review_count_string



# Function to extract Availability Status

def get_availability(soup):

  try:

    availability = soup.find("div", attrs={"id": "availability"})

    if availability is not None:

      availability_string = availability.get_text(strip=True)

    else:

      availability_string = ""

  except AttributeError:

    availability_string = ""

  return availability_string



if __name__ == "__main__":

  # Headers for request

  headers = {

    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/124.0.0.0 Safari/537.36",

    "Accept-Language": "en-US, en;q=0.5"

  }



  # The webpage URL

   # url = "https://www.amazon.in/Timex-Analog-Blue-Dial-Watch-TW00ZR262E/dp/B07H3K85H5/"

  url = "https://www.amazon.in/Timex-Analog-Blue-Dial-Watch-TW00ZR262E/dp/B07H3K85H5/ref=sr_1_1?dib=eyJ2IjoiMSJ9.w3D-pT1iHxJCViIQYFd1Q-eKm1FkhnRAMe2brGHMQP22YkAyFnBv31E5epvQzEnvUTA192PBtq8cBfKFhTmAFz_Ja86pWrEHaXIrbIh9SlJYqpchQuQzWw8b6_1qg1BvCvA2HgzAyQbPnnDgXHmSiGOhEzs3CFlA8cufde-pXKT7ke1jX0BAmdDkgUXgsHWyaQHDGoZjPedON5ahalFxR0z1ez_AePIvzYxBsokjnNXUPAQblwEREzppVJrKraUrNY4Lob5_Ph1wF0PaH4sMJRRYpo2xAwkxU-H_0ZckmLg.lwLMhAhsOt0yoypTKoBaTkZJsAWAtTa0kb45GOYssKg&dib_tag=se&pf_rd_i=2563504031&pf_rd_m=A1VBAL9TL5WCBF&pf_rd_p=17a14071-74d2-457f-a161-b8781ae22ce8&pf_rd_r=4C7ABT3T01QJ539RYYXY&pf_rd_s=merchandised-search-2&qid=1714379397&refinements=p_n_feature_seven_browse-bin%3A1480900031%2Cp_n_feature_fourteen_browse-bin%3A11142592031%7C27064188031%2Cp_72%3A1318476031%2Cp_89%3ACitizen%7CDANIEL+KLEIN%7CEmporio+Armani%7CFossil%7CGUESS%7CSeiko%7CSkagen%7CSonata%7CTIMEX%7CTommy+Hilfiger&s=watches&sr=1-1"

  

  # HTTP Request

  webpage = requests.get(url, headers=headers)



  # Soup Object containing all data

  soup = BeautifulSoup(webpage.content, "html.parser")



  # Function calls to display all necessary product information

  print("Product Title =", get_title(soup))

  print("Product Price =", get_price(soup))

  print("Product Rating =", get_rating(soup))

  print("Number of Product Reviews =", get_review_count(soup))

  print("Availability =", get_availability(soup))
```
